```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arquivos = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
```

```python
df = pd.read_csv(arquivos)
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigrantes do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao
```

```python
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_comparacao['Brasil'], label='Brasil', lw=2, color='yellow')
plt.plot(df_comparacao['Argentina'], label='Argentina', lw=2, color='blue')
plt.title('Imigrantes do Brasil e Argentina para o Canadá', fontsize= 16, loc='left')
plt.xlabel('Ano', fontsize = 14)
plt.ylabel('Número de imigrantes', fontsize = 14)
plt.xticks(['1980','1985','1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.grid(linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=2)
ax.set_title('Imigrantes do Brasil para o Canadá\n1980 a 2013', fontsize= 16, loc='left')
ax.set_xlabel('Ano', fontsize = 14)
ax.set_ylabel('Número de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.grid(linestyle='--')


plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigrantes do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigracao do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()



plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imagração dos quatro países da América do Sul para o Canadá de 1980 a 2013')

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')
axs[0, 0].set_xlabel('Ano')
axs[0, 0].set_ylabel('Número de imigrantes')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')
axs[0, 1].set_xlabel('Ano')
axs[0, 1].set_ylabel('Número de imigrantes')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')
axs[1, 0].set_xlabel('Ano')
axs[1, 0].set_ylabel('Número de imigrantes')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')
axs[1, 1].set_xlabel('Ano')
axs[1, 1].set_ylabel('Número de imigrantes')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul_ordenado = america_sul.sort_values(by='Total', ascending=True)
america_sul_ordenado
```

```python
cores = []
for pais in america_sul_ordenado.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('gray')


fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color= cores)
ax.set_title('Brasil é o quarto pais da América do Sul quando\no assunto é migração para o Canadá de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.get_xaxis().set_visible(False)
ax.set_frame_on(False)
ax.tick_params(axis='both', which='both', length=0)

for i , valor in enumerate(america_sul_ordenado['Total']):
    ax.text(valor + 20, i, str(valor), va='center', fontsize=10)

fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```
